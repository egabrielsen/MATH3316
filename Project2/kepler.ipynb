```python
import csv
import matplotlib.pyplot as plt

xfile=open('x.txt')
x = []
yfile=open("y.txt")
y = []
tfile=open("t.txt")
t = []

for row in csv.reader(xfile):
    x.append(row)
for row in csv.reader(yfile):
    y.append(row)
for row in csv.reader(tfile):
    t.append(row)

    
plt.plot(t, x, 'b', label="r")
plt.ylabel('time')
plt.xlabel('x')
plt.title('x vs t')
plt.show()
plt.plot(t, y, 'b', label="r")
plt.ylabel('time')
plt.xlabel('y')
plt.title('y vs t')
plt.show()
plt.plot(y, x, 'b', label="r")
plt.ylabel('y')
plt.xlabel('x')
plt.title('x vs y')
plt.show()
```
