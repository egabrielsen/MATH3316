```python
# Erik Gabrielsen
# Plotting Carbon
# MATH 3316

from pylab import *

time = loadtxt("time.txt")
temperature = loadtxt("Temp.txt")
c2mm = loadtxt("c2mm.txt")
c4mm = loadtxt("c4mm.txt")
c2mm800K = loadtxt("C2mm_800K.txt")
c2mm900K = loadtxt("C2mm_900K.txt")
c2mm1000K = loadtxt("C2mm_1000K.txt")
c2mm1100K = loadtxt("C2mm_1100K.txt")
c2mm1200K = loadtxt("C2mm_1200K.txt")
c4mm800K = loadtxt("C4mm_800K.txt")
c4mm900K = loadtxt("C4mm_900K.txt")
c4mm1000K = loadtxt("C4mm_1000K.txt")
c4mm1100K = loadtxt("C4mm_1100K.txt")
c4mm1200K = loadtxt("C4mm_1200K.txt")

figure(1)
imshow(c2mm)
colorbar(orientation='horizontal')
savefig("c2mm.png")

figure(2)
imshow(c4mm)
colorbar(orientation='horizontal')
savefig("c4mm.png")

figure(3)
plot(time, c2mm800K, label='800K')
plot(time, c2mm900K, label='900K')
plot(time, c2mm1000K, label='1000K')
plot(time, c2mm1100K, label='1100K')
plot(time, c2mm1200K, label='1200K')
xticks( (1*3600, 6*3600, 12*3600, 24*3600, 36*3600, 48*3600), ('1 hr', '6 hr', '12 hr', '24 hr', '36 hr', '48 hr') )
xlabel('Time (t)')
ylabel('C(x)')
title('Carbon Concentrations; Depth = 2mm ')
legend(loc='upper right')
savefig("c2mm_concentrations.png")


figure(4)
plot(time, c4mm800K, label='800K')
plot(time, c4mm900K, label='900K')
plot(time, c4mm1000K, label='1000K')
plot(time, c4mm1100K, label='1100K')
plot(time, c4mm1200K, label='1200K')

xticks( (1*3600, 6*3600, 12*3600, 24*3600, 36*3600, 48*3600), ('1 hr', '6 hr', '12 hr', '24 hr', '36 hr', '48 hr') )
xlabel('Time (t)')
ylabel('C(x)')
title('Carbon Concentrations; Depth = 4mm')
legend(loc='upper right')
savefig("c4mm_concentrations.png")
```
