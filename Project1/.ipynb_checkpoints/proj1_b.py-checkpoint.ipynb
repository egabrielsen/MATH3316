```python
import csv
import matplotlib.pyplot as plt

nfile=open('n.txt')
n = []
hfile=open("h.txt")
h = []
rfile=open("r_.txt")
r = []
Rfile=open("R.txt")
R = []

for row in csv.reader(nfile):
    n.append(row)
for row in csv.reader(hfile):
    h.append(row)
for row in csv.reader(rfile):
    r.append(row)
for row in csv.reader(Rfile):
    R.append(row)
    
plt.semilogy(n, r, 'b', label="r")
plt.semilogy(n, R, 'r--', label="R")
plt.ylabel('error')
plt.xlabel('n')
plt.title('R and r relating to n')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

```

```python
plt.loglog(h, r, 'b', label="r")
plt.loglog(h, R, 'r--', label="R")
plt.ylabel('error')
plt.xlabel('h')
plt.title('R and r relating to h')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
