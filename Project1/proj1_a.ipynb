```python
import csv
import matplotlib.pyplot as plt

zfile=open('z.txt')
z = []
f=open("f.txt")
cos = []
p4file=open("p4.txt")
p4 = []
p8file=open("p8.txt")
p8 = []
p12file=open("p12.txt")
p12 = []

for row in csv.reader(f):
    cos.append(row)
for row in csv.reader(p4file):
    p4.append(row)
for row in csv.reader(p8file):
    p8.append(row)
for row in csv.reader(p12file):
    p12.append(row)
for row in csv.reader(zfile):
    z.append(row)

plt.plot(z, cos, label="cos(x)")
plt.plot(z, p4, 'b--', label="p4")
plt.plot(z, p8, 'r', label="p8")
plt.plot(z, p12, 'g', label="p12")
plt.axis([-3, 3, -1, 1])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Taylor Polynomials of Cos(x)')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

zfile=open('z.txt')
z = []
err4file=open("err4.txt")
err4 = []
err8file=open("err8.txt")
err8 = []
err12file=open("err12.txt")
err12 = []

for row in csv.reader(err4file):
    err4.append(row)
for row in csv.reader(err8file):
    err8.append(row)
for row in csv.reader(err12file):
    err12.append(row)
for row in csv.reader(zfile):
    z.append(row)

plt.semilogy(z, err4, 'b--', label="err4")
plt.semilogy(z, err8, 'r', label="err8")
plt.semilogy(z, err12, 'g', label="err12")
plt.axis([-3, 3, -1, 1])
plt.ylabel('cos(x) - f(x)')
plt.xlabel('x')
plt.title('Error on Taylor Polynomial Series')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
